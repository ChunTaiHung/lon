```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import os
import matplotlib.pyplot as plt
from PIL import Image
import csv

if not os.path.exists("./images"):
    os.mkdir("./images")
```

```python
loan_url = "https://www.ptt.cc/bbs/Loan/index2963.html"
# headers 需得到 cookie
#ptt_cookie = {"over18": "1"}
```

```python
resp = requests.get(url=loan_url)
soup = BeautifulSoup(resp.text, "lxml")
print(soup.prettify())
```

```python
ptt_url = "https://www.ptt.cc"
article_links = {}
# local = []
localation = ["台北","新北","雙北","桃園","新竹"]
    
link_rent = soup.find_all("div",class_="r-ent")

for link in link_rent:
    try:
        link_a = link.find("a")
        link_author = link.find("div",class_="author")
        link_date = link.find("div",class_="date")
        if "個人信貸" in link_a.string:
            title = link_a.string
            url = ptt_url + link_a.get("href")
            resp2 = requests.get(url)
            soup2 = BeautifulSoup(resp2.text, "lxml")
            inform = soup2.find("div",class_="bbs-screen bbs-content")
            local2 = inform.contents[4].split("\n")[9].split("：")[-1]
            date = link_date.string
            if local2 in localation:
                article_links[title] = [date,author,url]

#             local.append(inform.contents[4])

    except:
        continue
article_links
```
